```python
 # Dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_path = "data.csv"
```

```python
cancer_df = pd.read_csv(data_path, encoding="ISO-8859-1")
```

```python
cancer_df.head()
```

```python
null_columns=cancer_df.columns[cancer_df.isnull().any()]
cancer_df[null_columns].isnull().sum()
```

```python
list(cancer_df)
```

```python
df = cancer_df.drop(['Unnamed: 32'], axis=1)
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df.head()
```

```python
df_final = df[['diagnosis','perimeter_mean','area_mean','compactness_mean','concave points_mean']]
df_final.head()
```

```python
X = df_final.drop("diagnosis", axis=1)
y = df_final["diagnosis"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=100, stratify=y)
#X_scaler = StandardScaler().fit(X_train)
#X_train_scaled = X_scaler.transform(X_train)
#X_test_scaled = X_scaler.transform(X_test)


# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=4))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
 # Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(
    X_train,
    y_train_categorical,
    epochs=60,
    shuffle=True,
    verbose=2
)
```
